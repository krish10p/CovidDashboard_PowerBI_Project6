For bringing city column into 2021_HUID.xls file, Used Python, Pandas for quick transformation:
steps:
1.	Loaded df1 : Confirmed positive cases of COVID19 in Ontario.csv
2.	Loaded df2 : 2021_HUID.xls excel_sheet
3.	Took unique_elements of df1:phu_id in s1

```python
file_path = r"C:\Users\krish\powerBI_Projects\CovidDashboard_PowerBI_Project6\raw_data\covid data\Confirmed positive cases of COVID19 in Ontario.csv"
df1 = pd.read_csv(file_path)
```

```python
df1.drop(columns=['Accurate_Episode_Date', 'Case_Reported_Date',
       'Test_Reported_Date', 'Specimen_Date', 'Age_Group', 'Client_Gender',
       'Outcome1'], inplace=True)
df1.head()
```

```python
# Specify the path to your Excel file
file_path = r"C:\Users\krish\powerBI_Projects\CovidDashboard_PowerBI_Project6\raw_data\covid data\Ontario_Marginalized_Index_2021.xlsx"

# Specify the sheet name you want to read
sheet_name = "2021_HUID"  # Replace with your actual sheet name

# Read the data from the specified sheet into a DataFrame
df2 = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
df2.head()
```

```python
s1 = df1.loc[:,'Reporting_PHU_ID'].unique()
s1
```

```python
# In dictionary dict1, received 34 pairs of phu_id:phu_city 
dict1 = {}

for j in s1:
    for i in range(len(df1)):
        if df1.iloc[i,1] == j:
            city = df1.iloc[i,4]
            dict1[j] = city   
            break   
#dict1
```

```python
df4 = pd.DataFrame(dict1,index=['PHU_City'])
df5 = df4.T
df5.reset_index(inplace=True)
df5['PHU_id'] = df5.loc[:,'index']
df5.drop(columns='index',inplace=True)
df5.head()
```

```python
# inner join in pandas:
df6 = pd.merge(df2, df5, left_on='HUID',right_on='PHU_id',how='inner')
df6.drop(columns='PHU_id',inplace=True)
df6.head()
```

```python
file_path = "2021_HUID_after_python_pd_transformation.csv"
df6.to_csv(file_path,index = False)
```
